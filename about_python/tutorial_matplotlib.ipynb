To get started with Matplotlib, you first need to import it. You can do this using the following code:

```python
import matplotlib.pyplot as plt
```

## Line Plot

The simplest and most commonly used plot is the line plot. To create a line plot, you need to provide two arrays - one for the x-axis and one for the y-axis. You can use the plot() function to create a line plot. Here's an example:

```python
import matplotlib.pyplot as plt

# Create the data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')

# Show the plot
plt.show()
```

## Scatter Plot
A scatter plot is another commonly used plot. To create a scatter plot, you need to provide two arrays - one for the x-axis and one for the y-axis. You can use the scatter() function to create a scatter plot. Here's an example:

```python
# Create the plot
plt.scatter(x, y)

# Add labels and a title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

## Bar Plot
A bar plot is useful when you want to show the distribution of a categorical variable. To create a bar plot, you need to provide an array for the x-axis and an array for the height of the bars. You can use the bar() function to create a bar plot. Here's an example:

```python
# Create the data
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 2, 5, 8]

# Create the plot
plt.bar(x, y)

# Add labels and a title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Plot')

# Show the plot
plt.show()
```

## Subplots
Subplots allow you to create multiple plots in the same figure. This is useful when you want to compare multiple sets of data or when you want to show the same data using different types of plots. To create subplots, you can use the subplots() function, which returns a figure object and an array of subplot objects. Here's an example:

```python
# Create the data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [10, 8, 6, 4, 2]

# Create the subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(7*2, 3))

# Create the first plot
axs[0].plot(x, y1, color="violet", linewidth=5, linestyle="--")
axs[0].set_xlabel('x-axis')
axs[0].set_ylabel('y-axis')
axs[0].set_title('Line Plot')

# Create the second plot
axs[1].scatter(x, y2, color="green", s=200, marker="*")
axs[1].set_xlabel('x-axis')
axs[1].set_ylabel('y-axis')
axs[1].set_title('Scatter Plot')

# figure title
fig.suptitle("Subplots")

# Show the plots
plt.show()
```

Here is another example

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```
