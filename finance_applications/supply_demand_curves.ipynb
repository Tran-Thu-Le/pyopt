```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
price = np.linspace(0, 100, 100)
supply = np.log(price + 1)
demand = np.log(101 - price)

# Finding the index of equilibrium point
idx = np.argwhere(np.diff(np.sign(supply - demand))).flatten()


# Plot data
plt.plot(price, supply, label='Supply')
plt.plot(price, demand, label='Demand')
plt.scatter(price[idx], supply[idx], color="blue", marker="*", s=200, label="equilibrium")
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.title('Supply-Demand Curves')
plt.legend()
plt.show()
```
